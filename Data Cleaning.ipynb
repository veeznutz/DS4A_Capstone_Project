## Heroin Related Overdose Death By County Data 2015- 2020 

```python
# Function to load a list of csv files 2015-2020
def load_dataset(name_path_dict):
    ''' Take in a dictionary with dataframe name and its CSV file paths and load into a dict of dataframes'''
    result_dict = {}
    for name, path in name_path_dict.items():
        result_dict[name] = pd.read_csv(path, skiprows=2)
    return result_dict
```

```python
path_dict = {
    "data_2015":"/home/jovyan/Capstone Project Stuff/Data/CA_Heroin-Related Overdose_Death_by County_2015_01.10.2022.csv",
    "data_2016":"/home/jovyan/Capstone Project Stuff/Data/CA_Heroin-Related Overdose_Death_by County_2016_01.10.2022.csv",
    "data_2017":"/home/jovyan/Capstone Project Stuff/Data/CA_Heroin-Related Overdose_Death_by County_2017_01.10.2022.csv",
    "data_2018":"/home/jovyan/Capstone Project Stuff/Data/CA_Heroin-Related Overdose_Death_by County_2018_01.10.2022.csv",
    "data_2019":"/home/jovyan/Capstone Project Stuff/Data/CA_Heroin-Related Overdose_Death_by County_2019_01.09.2022.csv",
    "data_2020":"/home/jovyan/Capstone Project Stuff/Data/CA_Heroin-Related Overdose_Death_by County_2020_01.09.2022.csv"}

# Load into dictionary of dataframes dfs
dfs = load_dataset(path_dict)
```

```python
# Check values in dfs
dfs["data_2015"]
```

```python
# Clean up the dataframes in dfs
# Delete last 2 rows and 95% LCL and 95% UCL column
for key, value in dfs.items():
    dfs[key] = dfs[key][["County","Rates","Counts"]]
    dfs[key].drop(dfs[key].index[-2:], inplace = True)
```

```python
dfs["data_2016"]
```

```python
# Add Year column into each dataframe
x = 2015
for key,value in dfs.items():
    dfs[key]["Year"] = x
    x += 1
```

```python
# Concatenate all year into one dataframe
all_df = pd.concat(dfs)
all_df
```

```python
type(all_df)
all_df.info()
```

```python
# Save into a CSV file for next step (cleaning, visualization)
all_df.to_csv("Data/Heroin_Deaths_by_County_2015-2020.csv", index=False)
```
